# Descriptive Statistics

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
```

```python
data=np.genfromtxt('PI-ANNA_DatenN=36CSV.csv', delimiter=";",skip_header=2,dtype=str)
type(data[1,1])

for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        data[i,j]=str(data[i,j]).replace(",",".")
        
data=data[:,1:11].astype('float')
```

```python
np.mean(data[15:39:,0:4])
```

```python
np.std(data[15:39:,0:4])
```

```python
plt.figure(figsize=(12.0,10.0))
plt.title('Mean expectation of pain; sick vs. healthy',fontsize=22)
index=np.arange(1)
bar_width=0.3
plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
plt.ylabel('Probability in %',fontsize=18)
plt.bar(index, np.mean(data[0:15:,0:4]), bar_width, yerr=np.std(data[0:15:,0:4]), color='red', label='Chronic pain patients')
plt.bar(index+bar_width, np.mean(data[15:39:,0:4]), bar_width, yerr=np.std(data[15:39:,0:4]), color='springgreen', label='Healthy controls')

plt.legend(fontsize=16)
plt.show()
```
