# Read data & sort by groups

```python
data=np.genfromtxt('PI-ANNA_DatenN=36CSV.csv', delimiter=";",skip_header=2,dtype=str)

type(data[1,1])

for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        data[i,j]=str(data[i,j]).replace(",",".")

data=data[:,1:5].astype("float")
```

```python
x=data
A=(x<25).astype('float')*20 + (x>58).astype('float')*70
A[A < .25] = 50
```

```python
allvpdata=[]
for vp in A:
    pain_values=[]
    for v in vp:
        if v==70.:
            pain_values.append('pain')
        elif v==50.:
            pain_values.append('avgpain')
        elif v==20.:
            pain_values.append('nopain')
    allvpdata.append(pain_values)
```

```python
sickdata=[]
controldata=[]
for vp in allvpdata[0:15]: 
    sickdata.append(vp)
for vp in allvpdata[16:35]:
    controldata.append(vp)
```

# Model Evidence Basic Model

```python
def buildbasicmodel():
        r=sp.variableNode(name='R', vrange=['pain', 'nopain', 'avgpain'])
        variable=[r]
        f=sp.freeEnergyFactorNode()
        f.addNeighbour(variable[0])
        factor=[f]

        def exprior(condrv,conditioners):
            if condrv[0] is None:
                return 10.0
            if condrv[0]=='nopain':
                return 0.3333
            if condrv[0]=='pain':
                return 0.3333
            if condrv[0]=='avgpain':
                return 0.3333
        f.setPrior(exprior)
        allNodes=[r,f]
        
        return(allNodes,variable,factor)
```

```python
def basicmodel_evidence(probdis_same=False,probdis_individual=False,sick=False,healthy=False,allvp=False):

    modelEvidence=0.0
    
    if probdis_same==True: # all patients' likelihood is from the same probability distribution
        if sick==True: 
            allNodes,variable,factor=buildbasicmodel()
            sp.startLearning(allNodes)
            for vp in sickdata:
                for v in vp:
                    variable[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(variable[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            modelEvidence+=factor[0].KLDiv_post_prior
            
        if healthy==True:
            allNodes,variable,factor=buildbasicmodel()
            sp.startLearning(allNodes)
            for vp in controldata:
                for v in vp:
                    variable[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(variable[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            modelEvidence+=factor[0].KLDiv_post_prior
            
        if allvp==True:
            allNodes,variable,factor=buildbasicmodel()
            sp.startLearning(allNodes)
            for vp in allvpdata:
                for v in vp:
                    variable[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(variable[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            modelEvidence+=factor[0].KLDiv_post_prior

        
    if probdis_individual==True:
        if sick==True:
            for vp in sickdata: # pro vp ein Model
                allNodes,variable,factor=buildbasicmodel()
                sp.startLearning(allNodes)
                for v in vp:
                    variable[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(variable[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                modelEvidence+=factor[0].KLDiv_post_prior
                    
                    
        if healthy==True:
            for vp in controldata: # pro vp ein Model
                allNodes,variable,factor=buildbasicmodel()
                sp.startLearning(allNodes)
                for v in vp:
                    variable[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(variable[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                modelEvidence+=factor[0].KLDiv_post_prior
                
        if allvp==True:
            for vp in allvpdata: # pro vp ein Model
                allNodes,variable,factor=buildbasicmodel()
                sp.startLearning(allNodes)
                for v in vp:
                    variable[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(variable[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                modelEvidence+=factor[0].KLDiv_post_prior
            
    return(modelEvidence)
```

```python
EvidenceBasicSick_Same=basicmodel_evidence(probdis_same=True,sick=True)
EvidenceBasicHealthy_Individual=(basicmodel_evidence(probdis_individual=True,healthy=True))
EvidenceBasicHealthy_Same=(basicmodel_evidence(probdis_same=True,healthy=True))
EvidenceBasicSick_Individual=(basicmodel_evidence(probdis_individual=True,sick=True))
```

# Model Evidence Latent Model


```python
def buildlatentmodel(causelistlen,effectlistlen):
    causelist=['L'+str(i) for i in range(causelistlen)]
    cause=[sp.variableNode(name, vrange=['morning','school','afternoon','eve']) for name in causelist]
    effectlist=['O'+str(i) for i in range(effectlistlen)]
    effect=[sp.variableNode(name, vrange=['pain','avgpain','nopain']) for name in effectlist]
    variables=[cause,effect]
    
    painprior=sp.freeEnergyFactorNode()
    painprior.addNeighbour(effect[0],isConditioner=False)
    painprior.addNeighbour(cause[0],isConditioner=True)
    
    factor=[painprior]

    def contextprior(condrv,conditioners):
        if condrv[0] is None: 
            return 10.0
        if conditioners[0]=='morning':
            if condrv[0]=='pain':
                return 0.333
            if condrv[0]=='nopain':
                return 0.333
            if condrv[0]=='avgpain':
                return 0.3333
        if conditioners[0]=='school':
            if condrv[0]=='pain':
                return 0.333
            if condrv[0]=='nopain':
                return 0.333
            if condrv[0]=='avgpain':
                return 0.3333
        if conditioners[0]=='afternoon':
            if condrv[0]=='pain':
                return 0.333
            if condrv[0]=='nopain':
                return 0.333
            if condrv[0]=='avgpain':
                return 0.3333
        else:
            if condrv[0]=='pain':
                return 0.333
            if condrv[0]=='nopain':
                return 0.3333
            if condrv[0]=='avgpain':
                return 0.3333

    painprior.setPrior(contextprior)
    allNodes=cause+effect+factor
    
    return (allNodes,cause,effect,factor)

# Umbenennen in cause and effect oder context oder sowas
```

```python
def latentmodelEvidence(probdis_equal=False,probdis_individual=False,sick=False,healthy=False,allvp=False):
    
    modelEvidence=0.0
    daytime=['morning','school','afternoon','eve']
    
    # Ein Model für alle VPs
    
    
    if probdis_equal==True:
        if sick==True:
            allNodes,cause,effect,factor=buildlatentmodel(1,1)
            sp.startLearning(allNodes)
            for vp in sickdata:
                for t,v in zip(daytime,vp):
                    cause[0].observe(t)
                    effect[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(effect[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            modelEvidence+=factor[0].KLDiv_post_prior
            
        if healthy==True:
            allNodes,cause,effect,factor=buildlatentmodel(1,1)
            sp.startLearning(allNodes)
            for vp in controldata:
                for t,v in zip(daytime,vp):
                    cause[0].observe(t)
                    effect[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(effect[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            modelEvidence+=factor[0].KLDiv_post_prior
            
            
        if allvp==True: 
            allNodes,cause,effect,factor=buildlatentmodel(1,1)
            sp.startLearning(allNodes)
            for vp in allvpdata:
                for t,v in zip(daytime,vp):
                    cause[0].observe(t)
                    effect[0].observe(v)
                    sp.runSumProduct(allNodes)
                    logp=np.log(effect[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            modelEvidence+=factor[0].KLDiv_post_prior
            
            
    # Jede VP kann potentiell eine eigene Probability distribution haben   
    if probdis_individual==True:
        if sick==True:
            for vp in sickdata: 
                allNodes,cause,effect,factor=buildlatentmodel(1,1)
                sp.startLearning(allNodes)
                for t,v in zip(daytime,vp):
                    effect[0].observe(v)
                    cause[0].observe(t)
                    sp.runSumProduct(allNodes)
                    logp=np.log(effect[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                modelEvidence+=factor[0].KLDiv_post_prior
                    
        if healthy==True:
            for vp in controldata: 
                allNodes,cause,effect,factor=buildlatentmodel(1,1)
                sp.startLearning(allNodes)
                for t,v in zip(daytime,vp):
                    effect[0].observe(v)
                    cause[0].observe(t)
                    sp.runSumProduct(allNodes)
                    logp=np.log(effect[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                modelEvidence+=factor[0].KLDiv_post_prior
                
        if allvp==True:
            for vp in allvpdata: 
                allNodes,cause,effect,factor=buildlatentmodel(1,1)
                sp.startLearning(allNodes)
                for t,v in zip(daytime,vp):
                    effect[0].observe(v)
                    cause[0].observe(t)
                    sp.runSumProduct(allNodes)
                    logp=np.log(effect[0].marginal()[-1])
                    modelEvidence+=logp
                    sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                modelEvidence+=factor[0].KLDiv_post_prior
        
                    
    return(modelEvidence)
```

# Model Evidence Markov chain

```python
def buildMarkovChainModel(responselistlen):
    factors=[]
    rlist=['R'+str(i) for i in range(responselistlen)]
    responses=[sp.variableNode(name, vrange=['pain','avgpain', 'nopain']) for name in rlist]
    
    r_prior=sp.freeEnergyFactorNode()
    r_prior.addNeighbour(responses[0])
    
    def r_prior_fct(condrv,conditioners):
        if condrv[0] is None:
            return 10.0
        if condrv[0]=='nopain':
            return 0.3333
        if condrv[0]=='pain':
            return 0.33333
        if condrv[0]=='avgpain':
            return 0.33333
        
    r_prior.setPrior(r_prior_fct)
    factors.append(r_prior)
    
    def chainprior(condrv,conditioners):
        if condrv[0] is None:
            return 10.0
        if conditioners[0]=='pain':
            if condrv[0]=='pain':
                return 0.33333
            if condrv[0]=='nopain':
                return 0.33333
            if condrv[0]=='avgpain':
                return 0.33333
        if conditioners[0]=='avgpain':
            if condrv[0]=='pain':
                return 0.33333
            if condrv[0]=='nopain':
                return 0.33333
            if condrv[0]=='avgpain':
                return 0.33333
        if conditioners[0]=='nopain':
            if condrv[0]=='pain':
                return 0.33333
            if condrv[0]=='nopain':
                return 0.33333
            if condrv[0]=='avgpain':
                return 0.33333
            
            
    chainparent=responses[0]
    for chainchild in responses[1:]:
        fe=sp.freeEnergyFactorNode()
        fe.addNeighbour(chainchild,isConditioner=False)
        fe.addNeighbour(chainparent,isConditioner=True)
        fe.setPrior(chainprior)
        factors.append(fe)
        chainparent=chainchild
        
    allNodes=responses+factors
    
    return(allNodes,responses,factors)
```

```python
def chainmodelEvidence(probdis_equal=False,probdis_individual=False, healthy=False, sick=False, allvp=False):
    
    
    modelEvidence=0.0
    
############################################################################################################################  
#################################### Alle 1 Modell #########################################################################

    if probdis_equal==True: 
        if sick==True: 
            allNodes,responses,factors=buildMarkovChainModel(4)               
            sp.startLearning(allNodes)
            for vp in sickdata:
                for r,v in zip(responses,vp):
                    r.observe(v)
                sp.runSumProduct(allNodes)
                logp=np.log(responses[0].marginal()[-1])
                modelEvidence+=logp
                sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            for f in factors:
                modelEvidence+=f.KLDiv_post_prior
            
        if healthy==True:
            allNodes,responses,factors=buildMarkovChainModel(4)               
            sp.startLearning(allNodes)
            for vp in controldata:
                for r,v in zip(responses,vp):
                    r.observe(v)
                sp.runSumProduct(allNodes)
                logp=np.log(responses[0].marginal()[-1])
                modelEvidence+=logp  
                sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            for f in factors: 
                modelEvidence+=f.KLDiv_post_prior
                
        if allvp==True: 
            allNodes,responses,factors=buildMarkovChainModel(4)               
            sp.startLearning(allNodes)
            for vp in allvpdata:
                for r,v in zip(responses,vp):
                    r.observe(v)
                sp.runSumProduct(allNodes)
                logp=np.log(responses[0].marginal()[-1])
                modelEvidence+=logp 
                sp.accumulateEvidence(allNodes)
            sp.learn(allNodes)
            for f in factors: 
                modelEvidence+=f.KLDiv_post_prior
            
        
################################################################################################################################ 
######################################## Individuelles Modell pro VP ##########################################################  
    
    if probdis_individual==True:
        if sick==True:
            for vp in sickdata:
                allNodes,responses,factors=buildMarkovChainModel(4)
                sp.startLearning(allNodes)
                for r,v in zip(responses,vp):
                    r.observe(v)
                sp.runSumProduct(allNodes)
                logp=np.log(responses[0].marginal()[-1])
                modelEvidence+=logp
                sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                for f in factors: 
                    modelEvidence+=f.KLDiv_post_prior

        if healthy==True:
            for vp in controldata:
                allNodes,responses,factors=buildMarkovChainModel(4)
                sp.startLearning(allNodes)
                for r,v in zip(responses,vp):
                    r.observe(v)
                sp.runSumProduct(allNodes)
                logp=np.log(responses[0].marginal()[-1])
                modelEvidence+=logp
                sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                for f in factors: 
                    modelEvidence+=f.KLDiv_post_prior
                    
                    
        if allvp==True:
            for vp in allvpdata:
                allNodes,responses,factors=buildMarkovChainModel(4)
                sp.startLearning(allNodes)
                for r,v in zip(responses,vp):
                    r.observe(v)
                sp.runSumProduct(allNodes)
                logp=np.log(responses[0].marginal()[-1])
                modelEvidence+=logp
                sp.accumulateEvidence(allNodes)
                sp.learn(allNodes)
                for f in factors: 
                    modelEvidence+=f.KLDiv_post_prior

    return(modelEvidence)
```

```python
modelEvidence=0.0
daytime=['morning','school','afternoon','eve']

allNodes,latent,observed,factor=buildlatentmodel(1,1)
sp.startLearning(allNodes)
for t,v in zip(daytime,sickdata[0]):
    latent[0].observe(t)
    observed[0].observe(v)
    sp.runSumProduct(allNodes)
    logp=np.log(observed[0].marginal()[-1])
    modelEvidence+=logp
    sp.accumulateEvidence(allNodes)
sp.learn(allNodes)
modelEvidence+=factor[0].KLDiv_post_prior

vp1_latent_same_evidence=modelEvidence
```
